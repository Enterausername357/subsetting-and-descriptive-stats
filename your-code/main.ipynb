#### Group the employees by number of years of employment. What are the average salaries in each group?

```python
# your code here
# your code here
#df.groupby(['Years']).mean(numeric_only=True).reset_index()
# Set numeric_only to True to only calculate the mean of numeric columns.
df.groupby(['Years'])['Salary'].mean().reset_index()
#trying out different ways to learn syntax, though I won`t remeber it hehe
```

####  What is the average salary per title?

```python
# your code here
#df.groupby(['Title']).mean(numeric_only=True).reset_index()
#by indicaating the '.mean(numeric_only=True' we only calculate the numeric columns,
#which, as I`ve noticed in the beginning are years & salary, quite handy!
df.groupby(['Title'])['Salary'].mean().reset_index()
#here`s a more specific way to do the same, for a case we had more numeric columns
```

####  Find the salary quartiles.


```python
# your code here
df['Salary'].quantile([0.25,0.5,0.75,1]).reset_index()
#I manualy added 1 to the quartiles function to see also the 4th one, the max.
```

#### Is the mean salary different per gender?

```python
# your code here
df.groupby(['Gender'])['Salary'].mean().reset_index()
#seems equal?
```

#### Find the minimum, mean and maximum of all numeric columns for each company department.



```python
# your code here
#df.groupby(['Department']).agg(['min','mean','max']).reset_index()
df.groupby(['Department'])[['Salary', 'Years']].agg(['min','mean','max']).reset_index()
#here`s a more specific way to do the same, for a case we had more numeric columns
```

#### Bonus Question:  for each department, compute the difference between the maximum and the minimum salary.
**Hint**: try using `agg` or `apply` combined with `lambda` functions.

```python
# your code here
df.groupby(['Department'])['Salary'].apply(lambda x: x.max() - x.min()).reset_index()
```
